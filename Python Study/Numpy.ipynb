### Concatenate - array 붙이기 

```python
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
np.array([1,2,3,4,5])
np.arange(1,6,1)
np.vstack((a,b))  #concatenate
np.hstack((a,b))
```

### Shape 가지고놀기 

```python
a=np.zeros([4,5])
a.shape
a.ndim
a.size
a.nbytes
a.dtype
np.ones_like(np.zeros([4,5]))  #Input의 shape을 그대로 따라한다.
```

### Slicing하기 

```python
a=np.arange(1,21,step=1).reshape([4,5])
print(a)
a[:,::2]   # x:y:z에서 x는 시작 index y는 끝 index, z는 step을 가르킨다.
a[::3,0:5:2]
```

```python
a=np.arange(1,11,1)
a
a[a>5]  #Boolean indexing
np.where(a>3)
a[np.where(a>3)] #Fancy indexing
a.take([1,4,7])
```

### reshape 생각해보기 

```python
a=np.arange(1,25,step=1)
a.reshape([2,12])
a.reshape([2,3,4])
b=a.reshape([2,3,4])
b[0,0,0]
b[0,0,3]
#a의 각 elements는 앞에서부터 차례대로 b의 가장 깊은 곳부터 채운다.
```
