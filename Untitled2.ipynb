```python
#이 부분은 2의 승수가 분모인 분수들을 둘러보기 위한 것이다, 그냥 참고용
import numpy as np
for i in [2,4,8,16,32,64]:
    x=[]
    for j in range(i):
        x.append(j/i)
    print('Scaling factor=%d'%i)
    print(x)
```

```python
target_level=63
x=np.array([i for i in range(-int(target_level/2),int(target_level/2)+1)])
bits=np.ceil(np.log(target_level) / np.log(2.0))
print("Target level equals %d, so bits equals %d"%(target_level,bits))
print(x/np.max(np.abs(x)))
x1 = (x/(np.max(np.abs(x))) * 0.5 + 0.5)
x2 = np.round(x1*(target_level-1))/2**(bits)
results=2*x2-(target_level-1)/2**(bits)
print("results\n",results)
index=(results*(2**(bits))+target_level-1)/2
print("index\n",index)
results2=(2*index-(target_level-1))/2**(bits)
print(results2)
```
